# LOAD PACKAGES

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define mz range
mz_min = 300
mz_max = 1200
n_features = 2001
mz = np.linspace(mz_min, mz_max, n_features)
```

```python
# Define num of peaks for each spectrum
num_peaks_min = 90
num_peaks_max = 300
num_peaks = np.random.randint(num_peaks_min,num_peaks_max)
```

```python
# Define the position of the peaks
peak_pos = np.random.randint(mz_min,mz_max,num_peaks)
```

```python
# Define the number of the spectra
num_spectra = 1000
```

```python
spectra = []
for i in range(num_spectra):
  spectrum = np.zeros(2001)
  for j in range(num_peaks):
    intensity = np.random.rand()
    mean = peak_pos[j]
    std = np.random.randn()
    spectrum += intensity*np.exp(-(mz-mean)**2/(2*std**2))
  spectra.append(spectrum)
```

```python
spectra = np.array(spectra)
plt.plot(mz,spectra.T)
plt.show()
```

```python
plt.imshow(np.corrcoef(spectra.T),cmap='coolwarm',vmin=-1,vmax=1)
plt.colorbar()
plt.show()
```
